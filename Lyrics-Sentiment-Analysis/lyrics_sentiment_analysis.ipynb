# 4. Sentiment Analysis:

```python
def get_sentiment(lyrics):
    blob = TextBlob(lyrics)
    sentiment = blob.sentiment.polarity
    return sentiment
```

```python
song_data = []

artists = ["Linkin Park", "Ed Sheeran", "Imagine Dragons", "The Weeknd", "Drake"]
for artist_name in artists:
    artist = genius.search_artist(artist_name, max_songs=3, sort="title")
    for song in artist.songs:
        sentiment_score = get_sentiment(song.lyrics)

        song_data.append([song.title, artist_name, round(sentiment_score, 4)])

df = pd.DataFrame(song_data, columns=['Title', 'Artist', 'Sentiment Score'])
df
```

## Using VADER Sentiment Analysis:

```python
analyzer = SentimentIntensityAnalyzer()

# Define the function to calculate sentiment score using VADER
def get_vader_sentiment(lyrics):
    sentiment = analyzer.polarity_scores(lyrics)
    return sentiment['compound']  # Compound score ranges from -1 (negative) to +1 (positive)

# Initialize an empty list to store the song details
song_data = []

artists = ["Linkin Park", "Ed Sheeran", "Imagine Dragons", "The Weeknd", "Drake"]
for artist_name in artists:
    artist = genius.search_artist(artist_name, max_songs=3, sort="title")
    for song in artist.songs:
        sentiment_score = get_vader_sentiment(song.lyrics)

        song_data.append([song.title, artist_name, round(sentiment_score, 4)])

# Create a DataFrame from the list
df_vader = pd.DataFrame(song_data, columns=['Title', 'Artist', 'Sentiment Score'])
df_vader
```

# Insights from Comparing the Two Sentiment Analysis Methods:
**Sentiment Range**:

- TextBlob (get_sentiment): The sentiment scores are in the range from -1 (negative) to +1 (positive), with scores generally close to 0 for more neutral tones (e.g., songs like "A.06" by Linkin Park and "100 GB (Un-Official)" by Drake, which have a score of 0.0).
- VADER (get_vader_sentiment): The VADER scores also range from -1 to +1, but the results show a wider spread. For example, VADER assigns high positive sentiment to songs like "2step" by Ed Sheeran with a score of 0.9945, but also gives negative scores like -0.9871 for "6 Inch Heel" by The Weeknd.

**Sensitivity to Song Lyrics**:

- TextBlob seems to provide more moderate sentiment scores, likely due to its simpler method of analysis. It often provides near-neutral scores (around 0), even for songs with relatively emotional content.
- VADER appears to be more sensitive to emotion in lyrics. It detects strong positive or negative sentiment more distinctly. For instance, "2step (American Remix)" by Ed Sheeran has a score of 0.9955 in VADER, indicating a strongly positive tone, whereas TextBlob assigns it only a modest 0.1206.

**Contrast in Song Sentiment**:

- Songs that both methods identify as positive include "2step" by Ed Sheeran, where VADER gives a very high score (0.9945) compared to TextBlob’s 0.1206. This could indicate that VADER is better at picking up on enthusiastic or positive sentiment in songs with energetic or upbeat tones.
- On the other hand, songs like "6 Inch Heel" by The Weeknd show a stark contrast: VADER assigns a strongly negative sentiment score (-0.9871), while TextBlob gives a mild negative score (-0.1962). This could be because VADER is more sensitive to intense negative emotion in the lyrics.

**Neutral Sentiment**:

- Both methods agree on neutral or almost neutral sentiment for songs like "A.06" by Linkin Park, which gets a score of 0.0 from both methods. This suggests that these methods are aligned when the sentiment is subtle or unclear.

**Genre and Artist Influence**:

- The differences in sentiment analysis scores might reflect the artists' emotional range or the lyrical content typical of their genre. For example, Drake's songs like "0 to 100 / The Catch Up" are strongly negative in VADER (-0.9987), which might be due to the intense, raw emotion often found in his lyrics. TextBlob still shows negative sentiment for these songs but to a lesser degree.
# Conclusion:
- **VADER** is generally more sensitive to extreme positive and negative emotions, making it suitable for analyzing songs with clear emotional extremes.
- **TextBlob** provides more moderate sentiment scores, making it more suitable for capturing nuanced or subtle feelings.
